```python
from matplotlib.image import imread
PATH = os.getcwd()
 
train_path_1 = PATH+'\\dataset\\training_set\\dogs'
train_path_2 = PATH+'\\dataset\\training_set\\cats'
train_batch_1 = os.listdir(train_path_1)
train_batch_2 = os.listdir(train_path_2)
x_train = []

for sample in train_batch_1:
    img_path = train_path_1 + '\\' + sample
    print(img_path)
    x = imread(img_path)
    x_train.append(x)
    
for sample in train_batch_2:
    img_path = train_path_2 + '\\' + sample
    print(img_path)
    x = imread(img_path)
    x_train.append(x)

# finally converting list into numpy array
x_train = pd.array(x_train)
# x_test = np.array(x_test)
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train/=255
X_test/=255
```

```python
number_of_classes = 10

Y_train = np_utils.to_categorical(y_train, number_of_classes)
Y_test = np_utils.to_categorical(y_test, number_of_classes)
```

```python
# Three steps to create a CNN
# 1. Convolution
# 2. Activation
# 3. Pooling
# Repeat Steps 1,2,3 for adding more hidden layers

# 4. After that make a fully connected network
# This fully connected network gives ability to the CNN
# to classify the samples

model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape=(28,28,1)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))

model.add(Conv2D(32, (3, 3)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3, 3)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))

model.add(Conv2D(64, (3, 3)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

# Fully connected layer
model.add(Dense(512))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(10))

model.add(Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
gen = ImageDataGenerator(rotation_range=8, width_shift_range=0.08, shear_range=0.3,
                         height_shift_range=0.08, zoom_range=0.08)

test_gen = ImageDataGenerator()
```

```python
train_generator = gen.flow(X_train, Y_train, batch_size=64)
test_generator = test_gen.flow(X_test, Y_test, batch_size=64)
```

```python
model.fit_generator(train_generator, steps_per_epoch=60000//64, epochs=5, 
                    validation_data=test_generator, validation_steps=10000//64)
```
